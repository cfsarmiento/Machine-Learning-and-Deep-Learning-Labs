```python
# Imports
import pandas as pd
import seaborn as sns
import numpy as np
import keras
from sklearn.model_selection import train_test_split
```

```python
# Import
bikeSharing = pd.read_csv('/content/train.csv')
```

```python
bikeSharing
```

```python
# Attributes of Interest
attributes = ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp',
              'humidity', 'windspeed', 'casual', 'registered']

# Initilize Model
model = keras.Sequential([

    keras.layers.Dense(100, activation='relu'),  # Input Layer, 100 perceptrons
    keras.layers.Dense(1, activation='softmax')  # Output Layer

])

# Compile Model
model.compile(optimizer='adam',
              loss='mse',
              metrics=['mean_squared_error'])
```

```python
# Split data into training/testing
x = bikeSharing.loc[:, attributes]
y = bikeSharing['count']

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
# Fit model
model.fit(xTrain, yTrain, epochs=50, batch_size=50)
loss, mse = model.evaluate(xTest, yTest)
print(f"Loss: {loss}, MSE: {mse}")
```
